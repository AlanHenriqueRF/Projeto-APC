# Projeto 2, implica em pesquisar as materias que o devido docente(prof) esta inscrito para dar aula, tendo em vista um arquivo no excel 

#### parametros, para criação das funções:
    * 'carga DOCENTE' basicamente, nos da o nome do docente, para consulta em tal arquivo, que devemos ler primeiro
    * 'disciplina D' neste pegamos o D como um inteiro, e ele filtra no arquivo, docentes com aquele valor ou maior, em ordem alfabetica
    * 'leia ARQ ' neste pegamos o arquivo .csv, excel, para ler-lo e comerarmos nosso codigo com base nisto 
    * 'matriculas COD1 COD2 ... CODN' lê os códigos das matérias, e filtra na planilha, para ver quantos matriculados tem
    * 'FIM' acaba o código
    

```python
def le_arquivo(dado):
    dado = dado.replace(' ','')
    with open(dado, 'r', encoding="utf-8") as arquivo:
        dados = input().split()
        dado = ''
        comando = dados[0]
        for i,index in enumerate(dados[1:]):
            dado += dados[i+1] + ' '

        if comando == 'carga':
            nome_prof = ''
            for info in dado:
                nome_prof += info
            print(f'{nome_prof}:')
        
        
            flag = True
            for i,linha in enumerate(arquivo):
                linha = linha.split(',') #lembrar de mudar para a ',', pois estou fazendo com ;

                if dado in linha[4]:
                    flag = False
                    dados_docente(linha)

            if flag == True:
                print(f'No hay {dado}...')

            
def dados_docente(dado): #pego o nome do professor, caso ja tenha sido chamado a planilha do excel
    if type(dado) is list:
        horario = dado[4][len(dado[4])-5:len(dado[4])-1]
        horario_atualizado = ''
        
        for j in horario:
            if j == '(' or j == ')':
                pass
            else:
                horario_atualizado += j
                
        print(f' * {dado[1]} ({dado[0]}):\n     Turma {dado[2]}: {horario_atualizado} ({dado[7]} alunos)')
        # print(f'[Carga total considerada: {120h} ({:.2f}h/aluno)]')
    else:
        dado = dado[-1].replace(' ','')
        print(f'No hay {dado}...')

        

def palavra_chave(comando):
    if 'carga' in comando:
        dados_docente(dado)
    elif 'matriculas' in comando:
        mostra_qnts_matriculados(dado)
    elif 'leia' in comando:
        le_arquivo(dado)
    elif 'disciplina' in comando:
        quantidade_de_docentes(int(dado))



while True:
    dados = input().split()
    palavras_que_pode = ['carga','disciplina', 'leia', 'matriculas']  
    if dados[0] not in palavras_que_pode:
        break
        
    dado = ''
    comando = dados[0]
    
    for i,index in enumerate(dados[1:]):
        dado += dados[i+1] + ' '
    
    palavra_chave(comando)
```
